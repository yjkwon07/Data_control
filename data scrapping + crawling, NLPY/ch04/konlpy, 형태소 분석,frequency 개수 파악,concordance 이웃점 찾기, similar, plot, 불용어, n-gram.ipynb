```python
global_unique = []
global_unique_count = []
```

```python
for row in text.split("\n"):
    term = t.nouns(row) #명사들만 줄 마다 
    unique = set(term)
    print(unique) # 중복 제외 
    global_unique += list(unique)
    global_unique = list(set(global_unique))
    global_unique_count.append(len(global_unique))
    print(len(unique),len(global_unique))
```

```python
#단일 형태소 찾기가 힘듬 고려대 (고려 + 대학교 ) 두개의 어절이 붙어있거나 복합명사 찾기 힘듬 
```

```python
def ngram(source, n):
    # 1-> 유니그램
    # 2 -> 바이그램
    # 3 -?트라이그램
    tokens = source.split()
    # 가 나 다 라 마 바 사 
    # 가, 나, 다, 라, 마, 바, 사
    #(n=2)가나, 나다, 다라, 라마, 마바,
    #(n=3) 가나다, 나다라, 다라마
    #보통 1음절은 버린다 .
    # 그리고 한 사람
    # 그리고한, 한사람(어절)
    # 고, _한, 한_, _사(음절) 한 글자 처리 길이가 하나인 단어를 처리 
    # 버리긴 한디 수사인 경우 일 -> 버리지 말아야, 어떻게 처리해야할까 
    # 유니그램이 아닌 바이그램으로 처리방법
    # 어디에 띄었나에 따라, 똑같은 1음절일지라도 다른곳에 카운팅 되도록
    
    for i in range(len(tokens)):
        tokens[i:i + n]
```

```python
stopwords = ["시발","씨발"] # 신조어는 품사정보가 없다stopwords = [(시발","NNG"),"씨발"] 
```

```python
rowString = "시발 련아 ^^"
```

```python
# 1차 쉬운방법
for arc in Kkma().morphs(rowString):
    if arc not in stopwords:
        print(arc)
```

```python
rowString = "시발련아^^"
```

```python
#2
for arc in Kkma().morphs(rowString):
    if arc not in stopwords:
        print(arc)
```

```python
def ngram(source, n=2):
    result = []
    for i in range(len(source)):
        result.append(source[i:i+n])
    return result
```

```python
ngram("가나다라")
```

```python
#2차 2음절 방법
for arc in Kkma().morphs(rowString):
    isStop =False
    
    for temp in ngram(arc):
        if temp in stopwords:
            isStop = True
    if not isStop:
        print(arc)
```
